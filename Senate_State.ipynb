## Senate Vote By State

This is general information about reporting votes to presidential race by state and its summary statistics.



```python
import pandas as pd

df = pd.read_csv('senate_state.csv')
df
```

```python
# No. of Senate vote records
count = df['state'].count()
print('No. of Senate vote records: ')
print(count)
```

```python
# Average no. of senate vote by state
avg = df['total_votes'].mean()
print('Average no. of votes by state: ')
print(avg.round(2))
```
